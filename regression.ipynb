```python
import pandas as pd
import numpy as np
import re
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/train-new-car-dataset.csv')
df.rename(columns={'Price':'Used_car_price','New_Price1':'Original_cost'},inplace=True)
df.head()
```

```python
data = df.copy()
data['Car_age'] = data.Year.apply(lambda x: 2023-x)
data.drop(['Unnamed: 0','Name','Variant'],axis=1,inplace=True)
data.drop(['New_Price','Year'],axis=1,inplace=True)
data.head()
```

```python
le = LabelEncoder()
data['Location'] = le.fit_transform(data['Location'])
data['Fuel_Type'] = le.fit_transform(data['Fuel_Type'])
data['Transmission'] = le.fit_transform(data['Transmission'])
data['Owner_Type'] = le.fit_transform(data['Owner_Type'])
data['Brand'] = le.fit_transform(data['Brand'])
data['Model'] = le.fit_transform(data['Model'])
data.head()
```
